```python
from ultralytics import YOLO
from pathlib import Path
```

```python
model = YOLO('../data/model.pt')
```

```python
results = model.predict(
    source='../data/test_images',          
    conf=0.75
    )
```

```python
output_folder = Path("test_detections")
output_folder.mkdir(parents=True, exist_ok=True)

for i, results in enumerate(results):
    # Save each image with detections
    save_path = output_folder / f"detected_{i:04d}.jpg"
    results.save(filename=str(save_path))
    print(f"Saved: {save_path.name}")
```

```python
from IPython.display import Image, display
import glob

for img_path in sorted(glob.glob("test_detections/*.jpg"))[:8]:
    display(Image(filename=img_path, width=700))
```
